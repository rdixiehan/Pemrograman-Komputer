```python
def input_1d_array():
    num_elements = int(input("Masukkan jumlah emlemen array: "))
    array = []

    for i in range(num_elements):
        element = float(input(f"Masukkan elemen ke-{i+1}: "))
        array.append(element)

        return array
    
array = input_1d_array()
print("Array yang dimasukkan:", array)
```

# Array Dimensi 2

```python
def input_2d_array():
    rows = int(input("Masukkan jumlah baris: "))
    cols = int(input("Masukkan jumlah kolom: "))

    matrix = []

    for row in range(rows):
        row_data = []
        for col in range(cols):
            element = float(input(f"Masukkan elemen [{row+1},{col+1}]: "))
            row_data.append(element)
        matrix.append(row_data)

    return matrix


# Example usage
matrix = input_2d_array()
print("Array 2D yang dimasukkan:", matrix)
```
